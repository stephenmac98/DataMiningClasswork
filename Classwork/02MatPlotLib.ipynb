# MatPlotLib

```python
x = np.arange(1, 100, 0.25)
y = np.sin(3*x/2)
plt.plot(x, y)
plt.show()
```

```python
#Without using show to separate, everything gets added to the same graph
x = np.arange(1, 100, 0.25)
y = np.sin(3*x/2)
z = np.sin(x*6)
plt.plot(x, y)
plt.plot(x, z)
```

```python
#Use show to put out graphs separately
x = np.arange(1, 100, 0.25)
y = np.sin(3*x/2)
z = np.sin(x*6)
plt.plot(x, y)
plt.show()
plt.plot(x, z)
plt.show()
```

```python
#Create legend
x = np.arange(1, 10, 0.25)
y = np.sin(3*x/2)
z = np.cos(x*6)
plt.plot(x, y, label='sin')
plt.plot(x, z, label='cos')
plt.legend(('sin', 'cos'))
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```
