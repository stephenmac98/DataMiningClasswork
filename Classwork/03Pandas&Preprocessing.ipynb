## Correlation Analysis

```python
pop_corr = pop_df[["hare", "lynx", "carrot"]].corr()
pop_corr
```

```python
# Handling label weirdness: https://stackoverflow.com/questions/3529666/

labels = ['Hare', 'Lynx', 'Carrot']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(pop_corr)
fig.colorbar(cax)

ax.set_xticklabels(['']+labels)
ax.set_yticklabels(['']+labels)

plt.show()
```
