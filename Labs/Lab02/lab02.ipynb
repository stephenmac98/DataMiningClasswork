```python
import numpy as np
import matplotlib.pyplot as plt

fileName = "lab02_populations.txt"

popTable = np.loadtxt(fileName, delimiter=',', skiprows=1)    
        
year = popTable[:, 0]
hare = popTable[:, 1]
lynx = popTable[:, 2]
carrot = popTable[:, 3]

yearHare = popTable[:, 0:2]
yearHareBig = yearHare[yearHare[:, 1] > 40000]

print("Cols: " + str(popTable.shape[1]) + " Rows: " + str(popTable.shape[0]))
print("Hare min: " + str(hare.min()) + " max: " + str(hare.max()) + " average:" + str(hare.mean()))
print("Lynx min: " + str(lynx.min()) + " max: " + str(lynx.max()) + " average:" + str(lynx.mean()))
print("Carrot min: " + str(carrot.min()) + " max: " + str(carrot.max()) + " average:" + str(carrot.mean()))
print("Years with hair pop over 40000" + str(yearHareBig))


plt.plot(year, hare, label='hare')
plt.plot(year, lynx, label='lynx')
plt.plot(year, carrot, label='carrot')
plt.legend(('hare', 'lynx', 'carrot'))
plt.ylabel('count')
plt.xlabel('year')
plt.show()

plt.hist(carrot, bins=8)
plt.ylabel("frequency")
plt.xlabel("carrots")
plt.show()

plt.scatter(hare, lynx)
plt.xlabel('hares')
plt.ylabel('lynxs')
plt.show()
```

Hare populations seem to be highes when the lynx population is lower, however there are a few extraneous data points including (40000, 60000) and (54000, 45000) that affect any conclusions. We also see a hich concentration where both populations are low, possibly hinting to other factors determining the population of both animals
